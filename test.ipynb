```python
import numpy
```

```python
a = numpy.zeros([3,2])
print(a)
```

```python
a[0,0] =1
a[0,1] =2
a[1,0] =9
a[2,1] =12
print(a)
```

```python
import matplotlib.pyplot
%matplotlib inline
```

```python
matplotlib.pyplot.imshow(a, interpolation="nearest")
matplotlib.pyplot.show()
```
